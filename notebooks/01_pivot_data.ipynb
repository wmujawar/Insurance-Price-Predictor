This notebook is used to created pivoted data, that will further be used by Tableau to generate heatmap of correlation

```python
import os
import numpy as np
import pandas as pd
```

```python
parent_dir = os.path.abspath(os.path.join(os.getcwd(), os.pardir))
raw_data_dir = os.path.join(parent_dir, 'data', 'raw')
processed_data_dir = os.path.join(parent_dir, 'data', 'processed')
```

```python
df = pd.read_csv(os.path.join(raw_data_dir, 'insurance.csv'))
```

```python
df.head()
```

```python
df = df.reset_index()
```

```python
df = df[['index', 'Age', 'Height', 'Weight', 'NumberOfMajorSurgeries', 'PremiumPrice']]
```

```python
df
```

```python
df_melted = pd.melt(df, id_vars='index', value_vars=['Age', 'Height', 'Weight', 'NumberOfMajorSurgeries', 'PremiumPrice'])
```
